```python
def find_root_of_word(word):
  stemmer = Stemmer()
  while True:
    if word[-3:] == "بان":    # like مهربان
      return word
    stem_word = stemmer.stem(word)
    if word == stem_word:
      return word
    word = stem_word
```

```python
def make_infinitive(verb):
  lemmatizer = Lemmatizer()
  lemmatized = lemmatizer.lemmatize(verb)
  split_lemmatized = lemmatized.split("#")
  if "#" not in lemmatized:
    return verb
  if split_lemmatized[0] != "":
    return split_lemmatized[0]+"ن"
  return split_lemmatized[1]
```

```python
names_keywords = dict()
verbs_keywords = dict()
VERB = "VERB"
NOT_WANTED_WORDS = ["كه", "مى", "نمى", "آنانكه", "يا", "نيز"]
NAMES = ["NOUN", "ADJ", "NOUN,EZ", "ADJ,EZ"]
NOT_WANTED = ["PUNCT", "ADP", "ADV", "ADP,EZ", "ADV,EZ", "SCONJ", "CCONJ"]
# ADVs and ADV,EZs are like "مگر", "چرا", "همانند", "آنچنان", "آنجا"
# ADPs and ADP,EZs are like "نزد", "در", "از", "تا", "به"
# SCONJs are like "تا", "چه", "اگر", "چون"
# CCONJ is like "و"
with open(FA_QURAN_FILE_NAME, 'r', encoding="utf-8") as fa_quran_file:
  line = fa_quran_file.readline()
  line_count = 1
  while line:
      fa_ayeh = line.split("|")[-1].strip()
      ayeh_word_tokenize = word_tokenize(fa_ayeh)
      first_word_tokenized = ayeh_word_tokenize[0]
      # When "و" is at the first of line it is detected as a NOUN. we don't want that.
      if first_word_tokenized == "و":
        ayeh_word_tokenize = ayeh_word_tokenize[1:]
      cur_pos_tags = tagger.tag(ayeh_word_tokenize)
      for word_pos_tag in cur_pos_tags:
        word = word_pos_tag[0]
        pos_tag = word_pos_tag[1]
        if pos_tag in NOT_WANTED or word in NOT_WANTED_WORDS:
          continue
        if pos_tag == VERB:
          word = make_infinitive(word)
          verbs_keywords[word] = 1
        else:
          names_keywords[word] = 1
      print("ayeh number:", line_count)
      line_count += 1
      line = fa_quran_file.readline()
```

```python
def preprocess_word(input_text):
    output_text = input_text.replace('ة', 'ه')
    output_text = output_text.replace('ي', 'ی')
    output_text = output_text.replace('ى', 'ی')     # replace 'ىا' with 'یا'
    output_text = output_text.replace('ؤ', 'و')
    output_text = output_text.replace('أ', 'ا')
    output_text = output_text.replace('إ', 'ا')
    output_text = output_text.replace('ك', 'ک')
    return output_text
```

```python
names_keywords_preprocessed = [preprocess_word(name) for name in names_keywords]
verbs_keywords_preprocessed = [preprocess_word(verb) for verb in verbs_keywords]
```

```python
def write_iterator_to_file(iterator, filename):
  # iterator can be a dictionary or a list or a tuple
  with open(filename, 'a') as f:
    for key in iterator:
      f.write(key+'\n')
```

```python
write_iterator_to_file(names_keywords_preprocessed, "drive/MyDrive/tmp/MscProject/quran_names_keywords.txt")
write_iterator_to_file(verbs_keywords_preprocessed, "drive/MyDrive/tmp/MscProject/quran_verbs_keywords.txt")
```
